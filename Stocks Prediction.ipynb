```python
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,train_size=0.7,random_state=11)
```

```python
for select k best algorith always select the score_function on the basis of this
Feature - Cat, Target - Cat ==> Chi-Square

Feature - Numeric, Target - Numeric ==> Correlation

Feature - Cat, Target - Numeric ==> f_regression

Feature - Numeric, Target - Cat ==> f_classification
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
```

```python
best_features = SelectKBest(score_func=f_classif, k=50)
fit = best_features.fit(X,Y)
```

```python
dd = pd.DataFrame({
    'Columns':list(X.columns),
    'Score':fit.scores_
     })
```

```python
dd.head(50)
```

```python
k_Xtrain=fit.transform(X_train)
k_Xtest= fit.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
lr= LogisticRegression()
```

```python
gg=lr.fit(k_Xtrain,Y_train)
```

```python
y_pred=lr.predict(k_Xtest)
```

```python
accuracy_score(Y_test,y_pred)
```

```python
confusion_matrix(Y_test,y_pred)
```

```python
from sklearn.feature_selection import RFE
```

```python
rfe= RFE(estimator=lr,step=1)
```

```python
rfe.fit(X_train,Y_train)
```

```python
selected_feature= pd.DataFrame({
    'Columns': list(X_train.columns),
    'Scoring': rfe.ranking_
})
```
